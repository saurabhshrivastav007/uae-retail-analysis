```python
import pandas as pd
```

```python
# Path to your Excel file
file_path = '../data/uae_retail_trade_2010_2016.xlsx'
```

```python
# Load the Excel file
df = pd.read_excel(file_path)
```

```python
# Preview the first 5 rows
df.head()
```

```python
df['Indicators'].unique()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# Shared in GDP 

```python
# Filter only this indicator
gdp_share = df[df['Indicators'] == 'Share in GDP at current prices (%)']

# Plot
plt.figure(figsize=(10, 5))
sns.lineplot(data=gdp_share, x='Date', y='Indicators_Values', marker='o', linewidth=2)
plt.title('Share in GDP at Current Prices (%) - Abu Dhabi (2010–2016)', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Percentage (%)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Insight 1: Share in GDP (%)**

From 2010 to 2013, the retail sector consistently contributed around 4.2% to Abu Dhabi’s GDP at current prices. This share slightly declined in 2014, possibly due to macroeconomic shifts. However, 2015 marked a strong rebound, with the sector crossing the 5% mark — a sign of renewed growth or structural expansion. By 2016, the sector stabilized at approximately 5.5%, reflecting steady performance and growing importance in the emirate’s economy.

```python
# Calculate Year-over-Year % Change
gdp_share = df[df['Indicators'] == 'Share in GDP at current prices (%)'].copy()
gdp_share['YoY Change (%)'] = gdp_share['Indicators_Values'].pct_change() * 100
gdp_share = gdp_share.sort_values(by='Date').reset_index(drop=True)
gdp_share
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(data=gdp_share[1:], x='Date', y='YoY Change (%)', palette='coolwarm', edgecolor='black')

plt.title('YoY Change in Retail Share of UAE GDP', fontsize=14)
plt.xlabel('Year')
plt.ylabel('YoY Change (%)')
plt.axhline(0, color='gray', linestyle='--')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

### KPI Insight: YoY Change in Retail Sector's GDP Share

From 2010 to 2014, the retail sector's contribution to the UAE's GDP saw a decline, hitting a low in 2014. However, in 2015, there was a significant YoY growth of nearly 36%, indicating a strong recovery or policy-driven boost. This momentum continued into 2016 with another 11% increase, reflecting a positive trend in the sector's economic impact.

Such shifts may suggest government initiatives, increased consumer spending, or post-crisis market recovery during that time.


# NON OIL GDP

```python
# Filter only this indicator
non_oil_gdp = df[df['Indicators'] == 'Share in non-oil GDP at current prices (%)']

# Plot
plt.figure(figsize=(10, 5))
sns.lineplot(data=non_oil_gdp, x='Date', y='Indicators_Values', marker='o', linewidth=2)
plt.title('Share in Non-oil GDP at Current Prices (%) - Abu Dhabi (2010–2016)', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Percentage (%)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Insight 2: Share in Non-Oil GDP (%)**

Between 2010 and 2014, the retail sector’s contribution to Abu Dhabi’s non-oil GDP showed a gradual decline — from 8.2% in 2010 to 7.8% in 2014. This trend suggests either slowdowns in retail or faster growth in other non-oil sectors. However, a sharp recovery occurred in 2015, with the share rebounding to 8.1%, likely linked to the same growth factors that drove the GDP rebound. The contribution then plateaued at this level in 2016, indicating a phase of stabilization.


```python
# Filter: Non-Oil GDP share
non_oil = df[df['Indicators'] == 'Share in non-oil GDP at current prices (%)'].copy()

# Sort and reset
non_oil = non_oil.sort_values(by='Date').reset_index(drop=True)

# Calculate YoY %
non_oil['YoY Change (%)'] = non_oil['Indicators_Values'].pct_change() * 100
non_oil
```

### KPI Insight: YoY Change in Retail Sector’s Share of Non-Oil GDP

Between 2010 and 2014, the retail sector’s share in the UAE’s non-oil GDP saw a gradual decline, indicating a slowdown in relative retail growth compared to other non-oil industries. However, 2015 marked a notable reversal with a ~3.85% YoY increase, followed by continued stability into 2016.

This trend mirrors the overall GDP pattern, suggesting that retail is a consistent contributor to diversification efforts beyond oil and gas.


# COMPENSATION EMPLOYEE (MILLION AED)

```python
import matplotlib.ticker as mtick

# Filter only this indicator
comp_emp = df[df['Indicators'] == 'Compensation of Employees (Million AED)']

# Plot
plt.figure(figsize=(10, 5))
sns.lineplot(data=comp_emp, x='Date', y='Indicators_Values', marker='o', linewidth=2)
plt.title('Compensation of Employees (Million AED) - Abu Dhabi (2010–2016)', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Million AED')
plt.grid(True)

# Format y-axis with commas
plt.gca().yaxis.set_major_formatter(mtick.StrMethodFormatter('{x:,.0f}'))

plt.tight_layout()
plt.show()
```

**Insight 3: Compensation of Employees (Million AED)**

The retail sector in Abu Dhabi saw a consistent increase in employee compensation between 2010 and 2016. Total wages rose from AED 8.2 billion in 2010 to over AED 14 billion by 2016 — marking a 70%+ increase over seven years. This growth suggests rising employment, improving salaries, or both. The trend reflects the sector's growing contribution to economic activity and its expanding workforce footprint in the emirate.


```python
# Filter for Compensation of Employees
comp_emp = df[df['Indicators'] == 'Compensation of Employees (Million AED)'].copy()

# Sort and reset index
comp_emp = comp_emp.sort_values(by='Date').reset_index(drop=True)

# Calculate YoY %
comp_emp['YoY Change (%)'] = comp_emp['Indicators_Values'].pct_change() * 100
comp_emp
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(data=comp_emp[1:], x='Date', y='YoY Change (%)', palette='coolwarm', edgecolor='black')

plt.title('YoY Change in Retail Share of UAE GDP', fontsize=14)
plt.xlabel('Year')
plt.ylabel('YoY Change (%)')
plt.axhline(0, color='gray', linestyle='--')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

### KPI Insight: YoY Change in Compensation of Employees (Million AED)

Employee compensation in the UAE retail sector remained relatively steady between 2010 and 2014. However, starting in 2015, there was a marked upward shift, with substantial year-over-year increases observed through 2016.

This trend may reflect a combination of sectoral growth, improved wages, or hiring expansions - all of which signal renewed investment and confidence in the retail labor force.


# Gross Output

```python
# Filter for Gross Output
gross_output = df[df['Indicators'] == 'Gross Output (% of GDP at current prices)'].copy()

# Sort and reset index
gross_output = gross_output.sort_values(by='Date').reset_index(drop=True)

# Calculate YoY %
gross_output['YoY Change (%)'] = gross_output['Indicators_Values'].pct_change() * 100
gross_output
```

### KPI Insight: YoY Change in Gross Output (% of GDP)

The retail sector’s gross output as a share of GDP remained relatively stable between 2010 and 2014. However, the period between 2014 and 2016 shows a resurgence — with an upward YoY trend indicating increased retail production and output.

This growth phase likely reflects expanded consumer demand, retail infrastructure development, or an increase in product/service offerings by the sector.


# Capital Formation

```python
# Filter for Capital Formation
capital_form = df[df['Indicators'] == 'Capital formation (% of GDP at current prices)'].copy()

# Sort and reset index
capital_form = capital_form.sort_values(by='Date').reset_index(drop=True)

# Calculate YoY %
capital_form['YoY Change (%)'] = capital_form['Indicators_Values'].pct_change() * 100
capital_form
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(data=capital_form[1:], x='Date', y='YoY Change (%)', palette='flare', edgecolor='black')

plt.title('YoY Change in Capital Formation (% of GDP)', fontsize=14)
plt.xlabel('Year')
plt.ylabel('YoY Change (%)')
plt.axhline(0, color='gray', linestyle='--')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

### KPI Insight: YoY Change in Capital Formation (% of GDP)

Capital formation in the retail sector — which reflects investment in assets and infrastructure — remained modest between 2010 and 2014. However, a strong upward spike in 2015 suggests significant investments were made, possibly to support expansion or modernization efforts.

This aligns with similar trends seen in compensation and output KPIs, indicating a broader retail sector revitalization during this period.


```python
# Select only the columns we need
columns = ['Date', 'Emirate', 'Indicators', 'Indicators_Values', 'YoY Change (%)']

# Combine all KPI DataFrames
final_df = pd.concat([
    gdp_share[columns],
    non_oil[columns],
    gross_output[columns],
    capital_form[columns],
    comp_emp[columns]
], ignore_index=True)

# Sort by Date and Indicator
final_df = final_df.sort_values(by=['Indicators', 'Date']).reset_index(drop=True)

# Preview
final_df.head()
```

```python
# Export the final DataFrame to CSV
final_df.to_csv('../data/uae_retail_kpi_cleaned.csv', index=False)
```
