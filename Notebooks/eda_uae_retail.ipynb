```python
import pandas as pd
```

```python
# Path to your Excel file
file_path = '../data/uae_retail_trade_2010_2016.xlsx'
```

```python
# Load the Excel file
df = pd.read_excel(file_path)
```

```python
# Preview the first 5 rows
df.head()
```

```python
df['Indicators'].unique()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Filter only this indicator
gdp_share = df[df['Indicators'] == 'Share in GDP at current prices (%)']

# Plot
plt.figure(figsize=(10, 5))
sns.lineplot(data=gdp_share, x='Date', y='Indicators_Values', marker='o', linewidth=2)
plt.title('Share in GDP at Current Prices (%) - Abu Dhabi (20102016)', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Percentage (%)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Insight 1: Share in GDP (%)**

From 2010 to 2013, the retail sector consistently contributed around 4.2% to Abu Dhabis GDP at current prices. This share slightly declined in 2014, possibly due to macroeconomic shifts. However, 2015 marked a strong rebound, with the sector crossing the 5% mark  a sign of renewed growth or structural expansion. By 2016, the sector stabilized at approximately 5.5%, reflecting steady performance and growing importance in the emirates economy.

```python
# Filter only this indicator
non_oil_gdp = df[df['Indicators'] == 'Share in non-oil GDP at current prices (%)']

# Plot
plt.figure(figsize=(10, 5))
sns.lineplot(data=non_oil_gdp, x='Date', y='Indicators_Values', marker='o', linewidth=2)
plt.title('Share in Non-oil GDP at Current Prices (%) - Abu Dhabi (20102016)', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Percentage (%)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Insight 2: Share in Non-Oil GDP (%)**

Between 2010 and 2014, the retail sectors contribution to Abu Dhabis non-oil GDP showed a gradual decline  from 8.2% in 2010 to 7.8% in 2014. This trend suggests either slowdowns in retail or faster growth in other non-oil sectors. However, a sharp recovery occurred in 2015, with the share rebounding to 8.1%, likely linked to the same growth factors that drove the GDP rebound. The contribution then plateaued at this level in 2016, indicating a phase of stabilization.


```python
import matplotlib.ticker as mtick

# Filter only this indicator
comp_emp = df[df['Indicators'] == 'Compensation of Employees (Million AED)']

# Plot
plt.figure(figsize=(10, 5))
sns.lineplot(data=comp_emp, x='Date', y='Indicators_Values', marker='o', linewidth=2)
plt.title('Compensation of Employees (Million AED) - Abu Dhabi (20102016)', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Million AED')
plt.grid(True)

# Format y-axis with commas
plt.gca().yaxis.set_major_formatter(mtick.StrMethodFormatter('{x:,.0f}'))

plt.tight_layout()
plt.show()
```

**Insight 3: Compensation of Employees (Million AED)**

The retail sector in Abu Dhabi saw a consistent increase in employee compensation between 2010 and 2016. Total wages rose from AED 8.2 billion in 2010 to over AED 14 billion by 2016  marking a 70%+ increase over seven years. This growth suggests rising employment, improving salaries, or both. The trend reflects the sector's growing contribution to economic activity and its expanding workforce footprint in the emirate.
